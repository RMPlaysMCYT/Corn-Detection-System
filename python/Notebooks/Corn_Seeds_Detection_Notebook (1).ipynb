```python
import tensorflow as tf
```

```python
import keras
```

```python
from keras._tf_keras.keras.preprocessing.image import ImageDataGenerator
```

```python
datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)
```

```python
!unzip "/usr/local/cornseeds/training_set.zip" -d "/usr/local/cornseeds/training_set"
```

```python
training_set1 = datagen.flow_from_directory(
    '/usr/local/cornseeds/training_set',
    target_size=(64, 64),
    batch_size=32,
    class_mode='binary'
)
```

```python
training_datagenerator1 = ImageDataGenerator(rescale = 1./255)
```

```python
testing_set1 = training_datagenerator1.flow_from_directory(
    '/usr/local/cornseeds/training_set',
    target_size=(64, 64),
    batch_size=32,
    class_mode='binary'
)
```

```python
cnn_process1 = tf.keras.models.Sequential()
```

# FIRST LAYER

```python
cnn_process1.add(
    tf.keras.layers.Conv2D(
        filters=32,
        kernel_size=3,
        activation='relu',
        input_shape=[64, 64, 3]
    )
)

cnn_process1.add(
    tf.keras.layers.MaxPool2D(
        pool_size = 2,
        strides = 2
    )
)
```

# SECOND LAYER

```python
cnn_process1.add(
    tf.keras.layers.Conv2D(
        filters=32,
        kernel_size=3,
        activation='relu',
    )
)
cnn_process1.add(
    tf.keras.layers.MaxPool2D(
        pool_size = 2,
        strides = 2
    )
)
```

```python
cnn_process1.add(tf.keras.layers.Flatten())
```

```python
cnn_process1.add(tf.keras.layers.Dense(units=128, activation='relu'))
cnn_process1.add(tf.keras.layers.Dropout(0.5))
```

```python
cnn_process1.add(tf.keras.layers.Dense(units=128, activation='sigmond'))
```

```python
cnn_process1.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
```

```python
cnn_process1.fit(x=training_set1, validation_data=testing_set1, epochs=20)
```
